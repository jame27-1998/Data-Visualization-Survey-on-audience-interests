```python
import numpy as np  
import pandas as pd
```

```python

df_sur = pd.read_csv('https://cocl.us/datascience_survey_data')
df_sur.head()
```

```python
df_sur.set_index('Unnamed: 0', inplace=True)
df_sur.index.name = None
df_sur.head()
```

```python
df_sur.sort_values(["Very interested"], axis=0,ascending=False, inplace=True) 
df_sur.head()
```

```python
df_sur['Very interested']=round((df_sur['Very interested'] / 2233)*100,2)
df_sur['Somewhat interested']=round((df_sur['Somewhat interested'] / 2233)*100,2)
df_sur['Not interested']=round((df_sur['Not interested'] / 2233)*100,2)
df_sur.head()
```

```python
ax = df_sur.plot(kind='bar', alpha=0.35, figsize=(20, 8),width=0.8, color=['#5cb85c', '#5bc0de', '#d9534f'])
```

```python
ax = df_sur.plot(kind='bar', alpha=0.35, figsize=(20, 8),width=0.8, color=['#5cb85c', '#5bc0de', '#d9534f'])

ax.set_title('Percentage of Respondents Interest in Data Science Area',fontsize=16)
ax.set_ylabel('percentage')
ax.set_xlabel('Data science topics')
ax.legend(fontsize = 14)
ax.set_frame_on(False)
ax.set_xlabel(False).set_visible(False)
ax.axes.get_yaxis().set_visible(False)
ax.set_xticklabels(('Data Analysis / Statistics','Machine Learning','Data Visualization','Big Data (Spark / Hadoop)','Deep Learning','Data Journalism'))
```
